```python
column_names = []
for i in tweets:
  column_names.append(i)
tweets_filtered = pd.DataFrame(columns = column_names)
```

```python
for idx, row in tweets.iterrows():
  if abs(row.toxic_original - row.toxic_noaccent) > 0.10:
    tweets_filtered = tweets_filtered.append(row)
```
